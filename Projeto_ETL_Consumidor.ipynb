```python
import requests

URL_DA_API = "https://mulch-ransack-previous.ngrok-free.dev/puxar-nomes"

def processo_etl():
    print("--- 📥 ETAPA: EXTRAÇÃO ---")
    try:
        # Faz a chamada para a sua API local via internet
        response = requests.get(URL_DA_API)

        # Verifica se a conexão deu certo
        if response.status_code == 200:
            dados = response.json()
            nomes = dados['usuarios']
            print(f"Sucesso! {len(nomes)} nomes extraídos da API.\n")

            print("--- ⚙️ ETAPA: TRANSFORMAÇÃO ---")
            for nome in nomes:
                # Saudação
                mensagem = f"Parabéns {nome}! Você concluiu o primeiro passo do seu projeto ETL."

                # 3. CARGA (Aqui apenas exibindo o resultado final)
                print(f"[CARGA]: {mensagem}")

            print("\n--- ✅ PROCESSO CONCLUÍDO ---")
        else:
            print(f"Erro na API: Status {response.status_code}")

    except Exception as e:
        print(f"Erro de conexão: {e}")
        print("Dica: Verifique se o Ngrok e o Uvicorn ainda estão rodando no seu PC.")

# Rodar o processo
processo_etl()
```
